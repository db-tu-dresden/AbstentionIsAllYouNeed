```python
from dash import Dash, html, dcc, callback, Output, Input, dash_table
import plotly.express as px
from plotly.subplots import make_subplots
import seaborn as sns
import pickle
import re
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

with open('pretrained_dict.pkl', 'rb') as f:
    pretrained_dict = pickle.load(f)    

with open('evaluation_full.pkl', 'rb') as f:
    df = pickle.load(f)
```

```python
spam = df[(df['Dataset'] == 'Spam_OOD') & (df['Method'] == 'Distilbert Rejector') & (df['Data_size'] == 10000)]
sns.lineplot(x='Coverage', y='Accuracy', data=pd.DataFrame(pretrained_dict["4_datasets_10k_Spam"]["accuracy_coverage"]), label='Distilbert trained on 4 tasks (10k)', color = '#648FFF')
sns.lineplot(x='Coverage', y='Accuracy', data=pd.DataFrame(pretrained_dict["4_datasets_2k_Spam"]["accuracy_coverage"]), label='Distilbert trained on 4 tasks (2k)', color = '#FFB000')
#sns.lineplot(x='Coverage', y='Accuracy', data=pd.DataFrame(pretrained_dict["Distil-AG_News_on_Spam"]["accuracy_coverage"]), label='Distilbert trained on AgNews', color = '#FE6100')

plt.axhline(y=spam[spam['Coverage'] == 1]['Accuracy'].tolist()[0], linestyle='--', color='black')
sns.lineplot(x=spam['Coverage'], y=spam['Accuracy'], label='Distilbert trained on Spam', color = '#DC267F')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.3), fontsize=14)
plt.xlabel('Coverage', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.gca().invert_xaxis()
plt.xlim(1, 0.5)
plt.savefig('appendix_exp4_4k_spam', dpi=300, bbox_inches='tight')  
plt.show()
```
